IMPORTING LIBRARIES

```python
import pandas as pd
import matplotlib.pyplot as mt
import numpy as np
```

IMPORTING DATASETS

```python
dataset=pd.read_csv('Salary_Data.csv')
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
```

```python
print(x)
```

```python
print(y)
```

SPLITTING DATASET INTO TRAINING SET AND TEST SET

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

training the simple linear regression model

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

PREDICTING THE RESULTS

```python
y_pred=regressor.predict(x_test)
```

VISUALISING THE TRAINING SET

```python
mt.scatter(x_train,y_train,color='red')
mt.plot(x_train,regressor.predict(x_train), color='blue')
mt.title('Salary vs Experience(training set)')
mt.xlabel('Years of Experience')
mt.ylabel('Salary')
mt.show()
```

VISUALISING THE TEST RESULTS

```python
mt.scatter(x_test,y_test,color='red')
mt.plot(x_train,regressor.predict(x_train), color='blue')
mt.title('Salary vs Experience(TEST set)')
mt.xlabel('Years of Experience')
mt.ylabel('Salary')
mt.show()
```
